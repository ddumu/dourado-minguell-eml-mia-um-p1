## Introducción

El problema del bandido de k-brazos presenta un agente que debe elegir secuencialmente entre k acciones donde cada una de ellas le ofrece una recompensa a partir de una distribución de probabilidad fija desconocida. El objetivo final del agente será maximizar la recompensa obtenida tras sus elecciones en un número finito de pasos. Las acciones posibles corresponden a cada uno de los bandidos (también conocidos como máquinas tragaperras) cuyos retornos están configurados mediante dichas distribuciones. Este planteamiento constituye un escenario es ideal para estudiar el dilema de la exploración y la explotación.

## Algoritmos

Mediante esta experimentación, se busca analizar y comparar los resultados de recompensas obtenidos por medio de algoritmos de diferentes familias de métodos (e-greedy, UCB y ascenso del gradiente) sobre bandidos con diferentes distribuciones
(Bernoulli, binomial y normal).

### Algoritmo $\epsilon$-Greedy

El algoritmo Greedy se basa en la premisa de elegir la acción con mejor rendimiento promedio hasta el momento. Inicialmente, se prueba, al menos una vez, cada acción y a partir de ese momento se elige siempre la acción con mejor promedio de recompensas. La variante $\epsilon$-Greedy añade un parámetro $\epsilon$ que indica la probabilidad de, en lugar de elegir la mejor acción, escoger una de forma aleatoria, añadiendo una componente de exploración a un algoritmo de explotación.

### Algoritmo UCB1

El algoritmo UCB1 trata la exploración de forma diferente a $\epsilon$-Greedy. En lugar de seleccionar una acción aleatoria con cierta probabilidad, UCB1 añade a la recompensa estimada de cada brazo un término de exploración que favorece a las acciones menos exploradas. De esta manera, el algoritmo sigue un enfoque optimista por la que un brazo menos explotado podría tener una recompensa media superior a la considerada óptima hasta el momento. Para controlar el nivel de exploración, se utiliza un pararámetro $c$ configurable.

### Algoritmo Softmax

En lugar de tener un parámetro $\epsilon$ que regula la probabilidad de explorar o una medida optimista para seleccionar acciones, el algoritmo Softmax asigna probabilidades a cada acción en función de sus valores de recompensa esperados. Por tanto, las acciones con mayor recompensa hasta el momento tendrán una mayor probabilidad de ser seleccionadas, aunque pueda ser seleccionada cualquiera de ellas. El algoritimo tiene un parámetro de temperatura, a mayor valor se le asigne más se parecerán las probabilidades y por tanto habrá más exploración.

## Distribuciones de probabilidad

### Distribución normal

Se define por una media y una desviación estándar. Es una distribución en forma de campana donde la mayoría de los valores se concentran alrededor de la media y disminuyen progresivamente hacia ambos extremos formando colas.

### Distribución binomial

Se usa para medir la probabilidad de obtener un número determinado de éxitos en una cantidad de ensayos independientes que solo tienen dos posibles resultados: éxito o fracaso.

### Distribución de Bernoulli

Esta es un caso particular de la distribución binomial donde el número de ensayos es 1.
