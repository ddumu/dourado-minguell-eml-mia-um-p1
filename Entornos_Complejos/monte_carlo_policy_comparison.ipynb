# Monte Carlo On-Policy vs Off-Policy para el entorno Taxi de Gymnasium

## Liberias

```python
import random
import numpy as np
import gymnasium as gym

from src.experiment import monte_carlo_every_visit
from src.plotting import plot_episodes_lengths, plot_episodes_rewards, plot_episodes_epsilons, plot_policy_episodes
```

## Constantes

```python
SEED = 42

N_EPISODES = 100_000
DISCOUNT_FACTOR = 0.99

EPSILON_MAX = 1.0
EPSILON_MIN = 0.1
EPSILON_DECAY_SLOWDOWN = 1

PLOT_MA_WINDOW = 1000
```

## Reproducibilidad

```python
random.seed(SEED)
np.random.seed(SEED)
```

## El entorno Taxi

### Espacio de acciones

```python
ACTIONS = set(range(0, 6))
print("Actions:", ACTIONS)
```

### Espacio de observaciones

```python
PASSENGER_LOCATIONS = set(range(0,5))
DESTINATIONS = set(range(0,4))

print("Passenger locations:", PASSENGER_LOCATIONS)
print("Destinations:", DESTINATIONS)
```

## Experimentación

### Ejecución

#### On-Policy

```python
env = gym.make("Taxi-v4")
ON_POLICY = True
Q_on_policy, episodes_rewards_on_policy, episode_lengths_on_policy, episodes_epsilons_on_policy = monte_carlo_every_visit(env, ON_POLICY, N_EPISODES, DISCOUNT_FACTOR, EPSILON_MAX, EPSILON_MIN, EPSILON_DECAY_SLOWDOWN, SEED)
```

#### Off-Policy

```python
env = gym.make("Taxi-v4")
ON_POLICY = False
Q_off_policy, episodes_rewards_off_policy, episode_lengths_off_policy, episodes_epsilons_off_policy = monte_carlo_every_visit(env, ON_POLICY, N_EPISODES, DISCOUNT_FACTOR, EPSILON_MAX, EPSILON_MIN, SEED)
```

### Evaluación

```python
plot_episodes_rewards(episodes_rewards_on_policy, "On-Policy", PLOT_MA_WINDOW, episodes_rewards_off_policy, "Off-Policy")
```

```python
plot_episodes_lengths(episode_lengths_on_policy, "On-Policy", PLOT_MA_WINDOW, episode_lengths_off_policy, "Off-Policy")
```

```python
plot_episodes_epsilons(episodes_epsilons_on_policy, "On-Policy", episodes_epsilons_off_policy, "Off-Policy")
```

```python
print("On-Policy")
plot_policy_episodes(Q_on_policy, 10)

print("Off-Policy")
plot_policy_episodes(Q_off_policy, 10)
```
