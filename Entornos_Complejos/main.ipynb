## El problema del Taxi

El problema del Taxi consiste en una cuadrícula de 5×5 en la que un taxi debe recoger a un pasajero y transportarlo hasta su destino. Tanto el origen como el destino del pasajero se encuentran en una de cuatro ubicaciones especiales del mapa. Al inicio de cada episodio, el pasajero aparece en una de estas posiciones y se le asigna como destino otra diferente. El objetivo es que el taxi se desplace por la cuadrícula, recoja al pasajero y lo deje en la ubicación indicada.

El taxi puede realizar seis acciones: desplazarse al norte, sur, este u oeste, recoger al pasajero y dejarlo en una ubicación. Cada movimiento tiene una penalización de −1 para fomentar rutas eficientes, mientras que las acciones de recogida o dejada realizadas de forma incorrecta reciben una penalización de −10. El episodio finaliza cuando el pasajero es dejado correctamente en su destino, obteniéndose una recompensa de +20.

Se trata de un problema de acciones discretas, ya que el agente únicamente puede elegir entre un conjunto finito de acciones predefinidas. Esta característica, junto con el número limitado de estados y acciones, lo convierte en un buen candidato para la aplicación de algoritmos tabulares como los métodos de Monte Carlo y algoritmos de diferencias temporales como SARSA y Q-Learning.

### Monte Carlo

Los métodos de Monte Carlo estiman el valor de laa acciones para los estados de la experiencia obtenida en cada episodio. Para ello, el algoritmo espera a que el episodio termine y actualiza sus estimaciones en base a la recompensa acumulada obtenida desde cada estado visitado.

Existen diferentes variantes, en este estudio se va a trabajar con on-policy y off-policy. En los métodos on-policy, la misma política utilizada para generar los episodios es la que se evalúa y mejora progresivamente. Por otro lado, los métodos off-policy permiten aprender una política objetivo diferente de la usada explorando el entorno mediante técnicas de muestreo por importancia.

### SARSA

SARSA es un algoritmo de diferencias temporales que, en lugar de esperar al final del episodio, actualiza las estimaciones del valor de las acciones de forma incremental en cada paso de interacción con el entorno. Estas actualizaciones se realizan utilizando la recompensa obtenida y la estimación asociada a la siguiente acción seleccionada por la política. Al mejorar la misma política que usa para generar las acciones, se considera un método on-policy.

### Q-Learning

Q-Learning es un algoritmo de diferencias temporales que, al igual que SARSA, actualiza las estimaciones del valor de las acciones de forma incremental durante la interacción con el entorno sin necesidad de esperar al final del episodio. Sin embargo, en lugar de utilizar la acción que la política actual tomaría en el siguiente estado, emplea la acción con el mayor valor estimado siguiendo una política greedy. De esta manera, Q-Learning aprende directamente una política óptima mientras explora el entorno mediante otra política, por lo que se considera un método off-policy.

## El problema de Mountain Car Continuous

El problema de MountainCarContinuous consiste en un vehículo situado en un valle unidimensional entre dos colinas. Al inicio de cada episodio, el coche aparece en el fondo del valle (en una posición inicial ligeramente aleatoria) con velocidad cero. El objetivo es que el vehículo ascienda la colina derecha y alcance una bandera situada en la cima. Sin embargo, el motor del coche carece de la potencia suficiente para superar la fuerza de la gravedad subiendo directamente, por lo que el agente debe aprender la física del entorno, balanceándose de un lado a otro para acumular la energía cinética y el impulso necesarios para coronar la pendiente.

En lugar de movimientos fijos, el coche cuenta con un espacio de acciones continuo representado por un único valor decimal que determina la fuerza y dirección del motor, oscilando entre -1.0 (máxima aceleración hacia la izquierda) y 1.0 (máxima aceleración hacia la derecha). Para incentivar la eficiencia energética, cada movimiento conlleva una pequeña penalización proporcional a la fuerza empleada (calculada como -0.1 multiplicado por el valor de la acción al cuadrado). El episodio finaliza de forma exitosa cuando el coche alcanza o supera la posición de la bandera (coordenada 0.45), momento en el que recibe una gran recompensa de +100, o bien se interrumpe si alcanza el límite máximo de pasos permitidos por el entorno.

Se trata de un problema de estados y acciones continuos, ya que tanto las métricas del entorno (posición y velocidad del coche) como las decisiones del agente admiten infinitos valores. Esta dimensionalidad hace imposible el uso directo de algoritmos tabulares o matrices de valores exactos. Por ello, este entorno resulta un candidato excelente para la aplicación de algoritmos de Aprendizaje por Refuerzo Profundo (Deep RL) orientados a espacios continuos, como DDPG, SAC o PPO, o bien para estudiar los desafíos de adaptar métodos discretos como Deep Q-Learning (DQN) mediante la discretización artificial de sus acciones.

### Sarsa Semi-Gradiente

Sarsa Semigradiente es una extensión del algoritmo de diferencias temporales que, en lugar de mantener una tabla discreta, utiliza aproximación de funciones (como redes neuronales o combinaciones lineales) para estimar el valor de las acciones en entornos continuos o muy grandes. Al igual que en su versión tabular, actualiza los parámetros (pesos) del modelo de forma incremental en cada paso de interacción. Estas actualizaciones se realizan ajustando los pesos en la dirección del gradiente del error, utilizando la recompensa obtenida y la estimación asociada a la siguiente acción seleccionada. Se denomina "semigradiente" porque ignora intencionadamente cómo cambian los pesos en el cálculo del objetivo (target), y al mejorar la misma política que usa para moverse, sigue considerándose un método on-policy.

### Deep Q-Learning

Deep Q-Learning es un algoritmo basado en diferencias temporales que utiliza redes neuronales profundas para estimar el valor de las acciones en entornos de alta complejidad. En lugar de actualizarse únicamente con la interacción inmediata, almacena sus vivencias en un búfer de memoria (Experience Replay) y utiliza lotes aleatorios de estas experiencias pasadas para actualizar los pesos de la red en cada paso, lo que estabiliza el aprendizaje. Estas actualizaciones se realizan utilizando la recompensa obtenida y la estimación de la mejor acción posible en el siguiente estado (el valor máximo), independientemente de la acción que se vaya a tomar después. Al aprender sobre la política óptima (buscando siempre el máximo valor) mientras utiliza una política diferente para explorar el entorno (como ϵ-greedy), se considera un método off-policy.
