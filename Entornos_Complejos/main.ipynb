## El problema del Taxi

El problema del Taxi consiste en una cuadrícula de 5×5 en la que un taxi debe recoger a un pasajero y transportarlo hasta su destino. Tanto el origen como el destino del pasajero se encuentran en una de cuatro ubicaciones especiales del mapa. Al inicio de cada episodio, el pasajero aparece en una de estas posiciones y se le asigna como destino otra diferente. El objetivo es que el taxi se desplace por la cuadrícula, recoja al pasajero y lo deje en la ubicación indicada.

El taxi puede realizar seis acciones: desplazarse al norte, sur, este u oeste, recoger al pasajero y dejarlo en una ubicación. Cada movimiento tiene una penalización de −1 para fomentar rutas eficientes, mientras que las acciones de recogida o dejada realizadas de forma incorrecta reciben una penalización de −10. El episodio finaliza cuando el pasajero es dejado correctamente en su destino, obteniéndose una recompensa de +20.

Se trata de un problema de acciones discretas, ya que el agente únicamente puede elegir entre un conjunto finito de acciones predefinidas. Esta característica, junto con el número limitado de estados y acciones, lo convierte en un buen candidato para la aplicación de algoritmos tabulares como los métodos de Monte Carlo y algoritmos de diferencias temporales como SARSA y Q-Learning.

### Monte Carlo

Los métodos de Monte Carlo estiman el valor de laa acciones para los estados de la experiencia obtenida en cada episodio. Para ello, el algoritmo espera a que el episodio termine y actualiza sus estimaciones en base a la recompensa acumulada obtenida desde cada estado visitado.

Existen diferentes variantes, en este estudio se va a trabajar con on-policy y off-policy. En los métodos on-policy, la misma política utilizada para generar los episodios es la que se evalúa y mejora progresivamente. Por otro lado, los métodos off-policy permiten aprender una política objetivo diferente de la usada explorando el entorno mediante técnicas de muestreo por importancia.

### SARSA

SARSA es un algoritmo de diferencias temporales que, en lugar de esperar al final del episodio, actualiza las estimaciones del valor de las acciones de forma incremental en cada paso de interacción con el entorno. Estas actualizaciones se realizan utilizando la recompensa obtenida y la estimación asociada a la siguiente acción seleccionada por la política. Al mejorar la misma política que usa para generar las acciones, se considera un método on-policy.

### Q-Learning

Q-Learning es un algoritmo de diferencias temporales que, al igual que SARSA, actualiza las estimaciones del valor de las acciones de forma incremental durante la interacción con el entorno sin necesidad de esperar al final del episodio. Sin embargo, en lugar de utilizar la acción que la política actual tomaría en el siguiente estado, emplea la acción con el mayor valor estimado siguiendo una política greedy. De esta manera, Q-Learning aprende directamente una política óptima mientras explora el entorno mediante otra política, por lo que se considera un método off-policy.
