# Monte Carlo On-Policy para el entorno Taxi de Gymnasium

## Liberias

```python
import random
import numpy as np
import gymnasium as gym

from src.experiment import monte_carlo_every_visit
from src.plotting import plot_episodes_lengths, plot_episodes_rewards, plot_episodes_epsilons, plot_policy_episodes
```

## Constantes

```python
SEED = 42

N_EPISODES = 100_000
ON_POLICY = True
DISCOUNT_FACTOR = 0.99

EPSILON_MAX = 1.0
EPSILON_MIN = 0.1
EPSILON_DECAY_SLOWDOWN = 1

PLOT_MA_WINDOW = 1000
```

## Reproducibilidad

```python
random.seed(SEED)
np.random.seed(SEED)
```

## El entorno Taxi

### Espacio de acciones

```python
ACTIONS = set(range(0, 6))
print("Actions:", ACTIONS)
```

### Espacio de observaciones

```python
PASSENGER_LOCATIONS = set(range(0,5))
DESTINATIONS = set(range(0,4))

print("Passenger locations:", PASSENGER_LOCATIONS)
print("Destinations:", DESTINATIONS)
```

## Experimentación

### Ejecución

```python
env = gym.make("Taxi-v4")
Q, episodes_rewards, episode_lengths, episodes_epsilons = monte_carlo_every_visit(env, ON_POLICY, N_EPISODES, DISCOUNT_FACTOR, EPSILON_MAX, EPSILON_MIN, EPSILON_DECAY_SLOWDOWN, SEED)
```

### Evaluación

```python
plot_episodes_rewards(episodes_rewards, "On-Policy", PLOT_MA_WINDOW)
```

```python
plot_episodes_lengths(episode_lengths, "On-Policy", PLOT_MA_WINDOW)
```

```python
plot_episodes_epsilons(episodes_epsilons, "On-Policy")
```

```python
plot_policy_episodes(Q, 10)
```
